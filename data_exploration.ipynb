```python
import pandas as pd
```

```python
books = pd.read_csv("data/books.csv")
books
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ax = plt.axes()
sns.heatmap(books.isna().transpose(), cbar=False, ax=ax)

plt.xlabel("Columns")
plt.ylabel("Missing Values")
```

```python
import numpy as np

books["missing_description"] = np.where(books["description"].isna(), 1, 0)
books["age_of_book"] = 2025 - books["published_year"]
```

```python
columns_of_interest = ["num_pages", "age_of_book", "missing_description", "average_rating"]
correlation_matrix = books[columns_of_interest].corr(method="spearman")

sns.set_theme(style="white")
plt.figure(figsize=(10, 10))
heatmap = sns.heatmap(correlation_matrix, annot=True, fmt=".2f")
heatmap.set_title("Correlation")
```

```python
books[(books["description"].isna()) |
      (books["num_pages"].isna()) |
      (books["average_rating"].isna()) |
      (books["published_year"].isna())
      ]
```

303 rows have missing values. Missing rows are less than 5% of the total dataset size. 
We can delete them

```python
book_missing = books[~(books["description"].isna()) &
      ~(books["num_pages"].isna()) &
      ~(books["average_rating"].isna()) &
      ~(books["published_year"].isna())
      ]
```

```python
book_missing
```

```python
book_missing["categories"].value_counts().reset_index().sort_values("count", ascending=False)
```

```python
book_missing["words_in_description"] = book_missing["description"].str.split().str.len()
```

```python
book_missing
```

```python
book_missing.loc[book_missing["words_in_description"].between(1, 4), "description"].count()
# there are 55 entries where the description is b/w 1 & 4 words long
# We can remove these entries
```

```python
book_missing.loc[book_missing["words_in_description"].between(5, 14), "description"].count()
```

```python
book_missing.loc[book_missing["words_in_description"].between(15, 25), "description"].count()
```

```python
book_missing_25_words = book_missing[book_missing["words_in_description"] >= 25]
```

```python
book_missing_25_words["title_and_subtitle"] = (np.where(book_missing_25_words["subtitle"].isna(), book_missing_25_words["title"],
                                                        book_missing_25_words[["title", "subtitle"]].astype(str).agg(": ".join, axis=1)))
```

```python
book_missing_25_words["tagged_description"] = book_missing_25_words[["isbn13", "description"]].astype(str).agg(" ".join, axis=1)
```

```python
book_missing_25_words.drop(["subtitle", "missing_description", "age_of_book", "words_in_description"], axis=1).to_csv("books_cleaned.csv", index=False)
```
